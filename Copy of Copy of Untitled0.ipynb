```python
import pandas as pd

# Load the data
downtime_factors = pd.read_csv('/content/drive/MyDrive/Colab DPEI/data/downtime factors.csv')
line_downtime = pd.read_csv('/content/drive/MyDrive/Combined New.csv')
manufacture_productivity = pd.read_csv('/content/drive/MyDrive/Colab DPEI/data/manufacture line productivity.csv')
product_info = pd.read_csv('/content/drive/MyDrive/Colab DPEI/data/product.csv')
# Melt the line_downtime dataframe to have a long format
line_downtime_long = line_downtime.melt(id_vars=['Batch'], var_name='Downtime Factor', value_name='Downtime Minutes')

# Merge with manufacture_productivity to get the operator for each batch
downtime_with_operator = pd.merge(line_downtime_long, manufacture_productivity[['Batch', 'Operator']], on='Batch')

# Sum the downtime minutes by operator
total_downtime_by_operator = downtime_with_operator.groupby('Operator')['Downtime Minutes'].sum().reset_index()
print(total_downtime_by_operator)
```

```python
# Convert Start Time and End Time to datetime
manufacture_productivity['Start Time'] = pd.to_datetime(manufacture_productivity['Start Time'])
manufacture_productivity['End Time'] = pd.to_datetime(manufacture_productivity['End Time'])

# Calculate the working time for each batch
manufacture_productivity['Working Time'] = (manufacture_productivity['End Time'] - manufacture_productivity['Start Time']).dt.total_seconds() / 60

# Sum the working time by operator
total_working_time_by_operator = manufacture_productivity.groupby('Operator')['Working Time'].sum().reset_index()
print(total_working_time_by_operator)
```

```python
# Merge total working time and total downtime
productivity_metrics = pd.merge(total_working_time_by_operator, total_downtime_by_operator, on='Operator', how='left')

# Fill NaN downtime with 0
productivity_metrics['Downtime Minutes'] = productivity_metrics['Downtime Minutes'].fillna(0)

# Calculate total time (working time + downtime)
productivity_metrics['Total Time'] = productivity_metrics['Working Time'] + productivity_metrics['Downtime Minutes']

# Calculate productivity ratio
productivity_metrics['Productivity Ratio'] = productivity_metrics['Working Time'] / productivity_metrics['Total Time']

print(productivity_metrics)
```

```python
# Merge downtime factors with line downtime
downtime_reasons = pd.merge(line_downtime_long, downtime_factors, left_on='Downtime Factor', right_on='Factor')

# Filter for operator errors
operator_error_downtime = downtime_reasons[downtime_reasons['Operator Error'] == 'Yes']

# Sum operator error downtime by operator
operator_error_downtime_by_operator = pd.merge(operator_error_downtime, manufacture_productivity[['Batch', 'Operator']], on='Batch')
operator_error_downtime_by_operator = operator_error_downtime_by_operator.groupby('Operator')['Downtime Minutes'].sum().reset_index()

print(operator_error_downtime_by_operator)
```
